```python
#break, continue, pass
```

```python
for i in range(5):
    if i == 3:
        # print(i)
        break
        
    print(i)
print("Loop ended")
```

```python
for i in range(5):
    if i == 3:
        continue
        
    print(i)
print("Loop ended")
```

```python
for i in range(5):
    if i == 3:
        pass
        
    print(i)
print("Loop ended")
```
