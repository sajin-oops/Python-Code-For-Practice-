```python
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

```python
fruits[1] = "mango"
print(fruits)
```

```python
fruits.append("graphs")
print(fruits)
```

```python
#Set - unordered,unique values only

numbers = {1,2,3,4,4,5}
print(numbers)
```

```python
numbers.add(6)
numbers
```

```python
numbers.remove(5)
numbers
```

```python
a = {1,2,3}
b = {3,4,5}
print(a | b)
print(a & b)
```
