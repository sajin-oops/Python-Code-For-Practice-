```python
squares = {x: x**2 for x in range(1,6)}
print(squares)
```

```python
names = ["Sajin","Snow","Milo"]
name_length = {name: len(name) for name in names}
print(name_length)
type(name_length)
```

```python
even_squares = {x: x**2 for x in range(1,11) if x % 2 == 0}
print(even_squares)
```

```python
result = {x: "even" if x % 2 ==0 else "odd" for x in range(2,6)}
print(result)
```

```python
keys = ["Name","Age","Role"]
values = ["Sajin",22,"Data Engineer"]
person = {k: v for k,v in zip(keys,values)}  #zip function acts like a physical zipper
print(person)
```

```python
data = {"a":"python","b":"sql","c":"kafka"}
upper_data = {k:v.upper() for k,v in data.items()}
upper_data
```

```python
table = {x: {y: x*y for y in range(1,4)} for x in range(1,4)}
print(table)
```
