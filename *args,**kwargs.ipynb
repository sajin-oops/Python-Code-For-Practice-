```python
# *args in python 
# *args allows a function to accept multiple positional arguments
# collects values in tuple
```

```python
def add(*args):
    print(args)
    return sum(args)

print(add(2,3))
print(add(5,3,6))
print(add(1,6,8))
print(add(1,1))
```

```python
# **kwargs in python
# **kwargs allows a function to accept multiple keyword arguments(key = value pairs)
# collects data into a dictionary
```

```python
def print_info(**kwargs):
    print(kwargs)

print_info(Name = "Sajin",Age = 23, Role = "Developer")
print_info(Name = "Cloudly", Company = "Google",Role = "Senior Data Engineer")
```

```python
# print(type(print_info))
print_info(name = "sajin",age = 23)
```

```python
def demo(*args,**kwargs):
    print("First_Value:",args)
    print("Second_Key_value:",kwargs)

demo(1,2,3,45,Name = "Sky")
```

```python
def train_model(model_name,*args,**kwargs):
    print("Model:",model_name)
    print("Extra Values:",args)
    print("Extra Settings",kwargs)

train_model("Random Forest",105,6,3,Max_depth = 5,random_sate = 42)
```
