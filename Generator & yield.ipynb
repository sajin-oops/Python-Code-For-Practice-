```python
# Generator & yield
# Generator is a special type of function that
# Uses yield instead of return 
```

```python
# Normal function vs generator
def numbers():
    return [1,2,3,4]
print(numbers())
```

```python
def numbers():
    yield 1
    yield 2
    yield 3

gen = numbers()
print(gen)
```

```python
# Now get values
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# return - ends function 
# yield - pauses function 
 
```

```python
def demo():
    print("Start")
    yield 1
    print("Middle")
    yield 2
    print("End")
    yield 3
generatee = demo()
```

```python
print(next(generatee))
```

```python
print(next(generatee))
```

```python
print(next(generatee)) 
```

```python
#
def big_generator():
    for i in range(10000):
        yield i

gen = big_generator()
print(next(gen))
print(next(gen))
```
