```python
#Reverse a string
text = 'Hello'
print(text[::-1])
```

```python
def reverse_bro(text):
    return text[::-1]

reverse_bro("Soo good")
```

```python
#Palindrome check
a = "HeH"
if a == a[::-1]:
    print("Yes Palindrome")
else:
    print("Nope its not")
```

```python
# Count vowels
text = input("Enter a string: ")
count = 0
for ch in text.lower():
    if ch in "aeiou":
        count += 1
print("Vowels:", count)
```
