```python
#Slicing and Indexing
```

```python
numbers = [10,20,30,40,50]
```

```python
print(numbers)
print(numbers[0])
print(numbers[1])
print(numbers[-1])
print(numbers[-2])
```

```python
text = "Python"
print(text[0])
print(text[-4])
print(text[2])
```

```python
numbers = [10,20,30,40,50]
print(numbers[1:4])
print(numbers[1:])
print(numbers[:])
```

```python
text = "Python"
print(text[::2])
print(text[::3])
```

```python
text = "Python"
print(text[-3:])
print(text[-5:])
```

```python
numbers = [10,20,30,40,50]
print(numbers[-4:-2])
print(numbers[-4:-1])
```
