```python
# Decorator in python


# A decorator is a function that takes another function.
# adds extra behavior
# returns a new modified function
# without changing the original function code.
```

```python
def greet():
    print("Hello there")

print(greet)
print(type(greet))
greet()
```

```python
say_hi = greet
say_hi()
```

```python
def outer():
    def inner():
        print("Inside inner function")
    inner()

outer()
```

```python
def outer():
    def inner():
        print("Hello from inner")
    return inner

new_function = outer()
new_function()
```

```python
#creating first decorator
def my_decorator(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper

def say_hello():
    print("Hello Sajin!")


decorated = my_decorator(say_hello)
decorated()
```

```python
# Using @ Syntax (real Decorator Style)
# instead of manually doing something like this (decorated = my_decorator(say_hello))

@my_decorator
def say_hello():
    print("Hello Sajin ak!")
say_hello()
```

```python
def my_decorator_bro(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper

@my_decorator_bro
def say_hi():
    print("Hi Bro!")
say_hi()
```

```python
def greet(name):
    print("Hello",name)

greet("Sky")
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before execution")
        func(*args, **kwargs)
        print("After execution")
    return wrapper

@my_decorator
def greet(name):
    print("Hello", name)

greet("Sajin")
```

```python
def login_required(func):
    def wrapper(user):
        if user == "admin":
            return func(user)
        else:
            print("Access Denied")
    return wrapper

@login_required
def dashboard(user):
    print("Welcome to dashboard", user)

dashboard("admin")
dashboard("Guest")
```
