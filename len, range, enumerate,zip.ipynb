```python
#len()
names = ["Sky","Snow","White"]
print(names)
print(len(names))
```

```python
#range
for i in range(1,4):
    print(i)
```

```python
for i in range(0,7,2):
    print(i)
```

```python
#enumerate()   - index + value together
names = ["sajin","white","cloud"]
for index, name in enumerate(names):
    print(index,name)
```

```python
for index,name in enumerate(names,start = 1): #here name - iterable , start - defult is 0 but we can change it
    print(index,name)
```

```python
#zip  - pairs elements position wise 

names = ["sajin","white","cloud"]
scores = [89,85,99]
for name,score in zip(names,scores):
    print(name,score)
```

```python
names = ["sajin","white","cloud"]
scores = [89,85,99]

for i, (name,score) in enumerate(zip(names,scores),start = 1):
    print(i,name,score)
```
