```python
#type hints & annotations
```

```python
# Variables without type hints
age = 23  
name = "SAJIN"

print(age)
print(name)
print(type(age))
print(type(name))
```

```python
# Variables with type hints
age: int = 23
name: str = "SAJIN"
print(age)
print(name)
print(type(age))
print(type(name))
```

```python
# With type hints
x: int = 33
y: float = 3.4
is_active = bool = True

print(x)
print(type(x))
print(y)
print(type(y))
print(is_active)
print(type(is_active))
```

```python
#with function
def add(a: int, b:int)-> int:
    return a + b

result  = add(3,3)
print(result)
```

```python
def greet(name: str) -> str:
    return f"Hello {name}"

greet("Sajin")
```

```python
greet("Cloudy")
```

```python
from typing import List
def total_marks(marks: List[int])-> int:
    return sum(marks)


#calling that function
total_marks([80,70,90])
```

```python
total_marks([66,67,64,56,67])
```
