```python
#List comprehensions
```

```python
numbers = []
for i in range(5):
    numbers.append(i)

print(numbers)
```

```python
numbers = [i for i in range(6)]
print(numbers)
```

```python
type(numbers)
```

```python
num = [i for i in range(5)]
print(num)
```

```python
squares = [i*i for i in range(1,6)]
print(squares)
```

```python
evens = [ ]
for i in range(1,11):
    if i % 2 ==0:
        evens.append(i)
```

```python
evens
```

```python
evens = [i for i in range(1,21) if i % 2 == 0]
```

```python
evens
```

```python
result = ["Even" if i % 2 ==0 else "Odd" for i in range(1,8)]
print(result)
```

```python
names = ["sajin","abishek","aaron","lijin"]
upper_names = [name.upper() for name in names]
print(upper_names)
```

```python
names = ["SajIN","AaROn"]
lower_case = [name.lower() for name in names]
print(lower_case)
```

```python
letters = [ch for ch in "SnowGold"]
print(letters)
```

```python
pairs = []
for x in [1,2]:
    for y in [3,4]:
        pairs.append((x,y))
print(pairs)
```

```python
pair = [(x,y) for x in [1,2] for y in [3,4]]
print(pair)
```

```python
salaries = [45000,34000,56000]
high_salary = [s for s in salaries if s > 55000] 
print(high_salary)
```
